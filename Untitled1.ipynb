```python
import pandas as pd

df = pd.read_csv(r"C:\Users\SUBHA PRAYASH\Downloads\Combined_dataset.csv")

print("First 5 Rows:")
print(df.head())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

df = df.drop_duplicates()

df["rating_plus_reviews"] = df["rating"] + df["ratings_count"]

filtered_df = df[df["rating"] > 4.0]

print("\nFiltered Data:")
print(filtered_df[["title", "rating"]].head())

df.to_csv("cleaned_combined_dataset.csv", index=False)

print("\nDataset cleaned and saved successfully!")
```
